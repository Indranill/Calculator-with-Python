```python
from tkinter import *
root=Tk()
root.title("calculator")
txtdisp=Entry(root,font="arial",width=30,justify=RIGHT)
txtdisp.grid(row=0,column=0,columnspan=8,pady=5,padx=1)

num="789456123"

i=0
btn=[]
def number(a):
    crr=txtdisp.get()
    txtdisp.delete(0,END)
    txtdisp.insert(0,crr+a)

for j in range(3,6):
    for k in range(0,3):
        btn.append(Button(root,width=6,height=2,font="arial",bd=4,text=num[i]))
        btn[i].grid(row=j,column=k,pady=1,padx=2)
        btn[i]["command"]=lambda x=num[i]:number(x)
        i+=1
        
        
sign="x/+-%="
l=0

    
def eq():
    second_number=txtdisp.get()
    txtdisp.delete(0,END)
    if(sign=="-"):
        txtdisp.insert(0,f_num-int(second_number))
    elif(sign=="+"):
        txtdisp.insert(0,f_num+int(second_number))
    elif(sign=="x"):
        txtdisp.insert(0,f_num*int(second_number))
    elif(sign=="%"):
        txtdisp.insert(0,f_num%int(second_number))
    elif(sign=="/"):
        txtdisp.insert(0,f_num%int(second_number))
    else:
        txtdisp.insert(0,"0")
    return 0
def calc(a):
    first_num=txtdisp.get()
    txtdisp.delete(0,END)
    global f_num
    f_num=int(first_num)
    global sign
    sign=a
    
    
    
for j in range(3,6):
    for k in range(3,5):
        btn.append(Button(root,width=6,height=2,font="arial",bd=4,text=sign[l]))
        btn[i].grid(row=j,column=k,pady=1,padx=2)
        btn[i]["command"]=lambda x=sign[l]:calc(x)
        l+=1
        i+=1
        
btn[14]["command"]=lambda:eq()     
def clear():
    txtdisp.delete(0,END)
clrbtb=Button(root,width=6,height=2,font="arial",bd=4,text="Clr",command=lambda:clear())
clrbtb.grid(row=6,column=0,pady=3,padx=3)
root.mainloop()
```
